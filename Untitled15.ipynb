```python

def gender_onehot(X):
    
    for i in range(X.shape[0]):
        if('male' in X[i]  ):
            X[i,2]=1
        elif('female' in X[i]  ):
            X[i,2]=0
        
    
    return X
    
```

```python
def embarked_onehot(X):
    
    for i in range(X.shape[0]):
        if('Q' in X[i]  ):
            X[i,7]=0
        elif('S' in X[i]  ):
            X[i,7]=1
        elif('C' in X[i]  ):
            X[i,7]=2
       
        
    
    return X
```

# Get training and test set for predicting age from both training and test data

```python
count=0
for i in range(X_tr3.shape[0]):
    if(np.isnan(X_tr3[i,3])):
      count+=1 
print(count)
    
```

```python
X_agetr=[]
agetr=[]
X_agets=[]
agets=[]
print(X_tr3.shape)
for i in range(X_tr3.shape[0]):
    m=[X_tr3[i,0],X_tr3[i,1],X_tr3[i,2],X_tr3[i,4],X_tr3[i,5],X_tr3[i,6],X_tr3[i,7]]
    
    if((np.isnan(X_tr3[i,3]))==False):
       
        X_agetr.append(m)
        agetr.append(X_tr3[i,3])
    else:
        X_agets.append(m)
        
        
        
X_agetr=np.array(X_agetr)
agetr=np.array(agetr)

X_agets=np.array(X_agets)
```

```python
import  sklearn.model_selection 
import tqdm.notebook
from sklearn.ensemble import GradientBoostingRegressor as gbr
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
```

```python
import warnings
warnings.simplefilter('ignore')
```

```python
scaling = StandardScaler()
scaling.fit(X_agetr)
Xtr_age = scaling.transform(X_agetr)
print(len(X_agetr))
Xts_age = scaling.transform(X_agets)
```

```python
learning_rate1=np.arange(0.1,1,0.045)
n_estimators1=np.arange(100,5000,245)
print(len(learning_rate1))
print(len(n_estimators1))

#loss1=['ls', 'lad', 'huber', 'quantile']
nfold=10
kf = KFold(n_splits=nfold)
kf.get_n_splits(Xtr_age)
acc=np.zeros((20,nfold))
pbar = tqdm.notebook.tqdm(
    total=nfold*20, initial=0,
    unit='fits', unit_divisor=20, desc='Model order test')
for i in range(20):
    for ifold, I in enumerate(kf.split(Xtr_age)):
        Itr, Its = I
        Xtr = Xtr_age[Itr]
        Xts = Xtr_age[Its]
        ytr = agetr[Itr]
        yts = agetr[Its]
        scaling = StandardScaler()
        scaling.fit(Xtr)
        Xtrs = scaling.transform(Xtr)
        reg=gbr(loss='quantile', learning_rate=learning_rate1[i],n_estimators=n_estimators1[i])
        reg.fit(Xtr,ytr)
        yhat=reg.predict(Xts)
        print(yts.shape)
        print(yhat.shape)
        acc[i,ifold]= r2_score(yts, yhat) 
        pbar.update(1)
```

```python
print(acc)
```

```python
rsq_mean = np.mean(acc,axis=1)
rsq_se = np.std(acc,axis=1)/np.sqrt(nfold-1)
plt.errorbar(learning_rate1, rsq_mean, rsq_se)
plt.grid()
plt.xlabel('learning rate')
plt.ylabel('Test R^2')
```
